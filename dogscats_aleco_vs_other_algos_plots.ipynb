```python
%matplotlib inline

from matplotlib import pyplot
import matplotlib.pyplot as plt
from pylab import genfromtxt
import numpy as np
```

# Visualize Acc. & Loss#

```python
##############################################################################

Optimizers = ["ALECO2" , "RMSprop" , "Adam" , "Nadam" , "SGD" , "SGDMo" , "SGDNes" , "SGDMoNes"]

##############################################################################
```

```python
# Parameters for ALECO2:
# dP = 0.01
# xi = 0.85
```

```python
plt.figure(figsize=(15, 10))

for optimizer in Optimizers:
    pyplot.plot(genfromtxt("./dogscats_results/loss_"+str(optimizer)+".txt"), label = str(optimizer))

plt.title('loss')  
plt.ylabel('loss')  
plt.xlabel('epoch')  
pyplot.legend();
pyplot.show();
```

```python
plt.figure(figsize=(15, 10))

for optimizer in Optimizers:
    pyplot.plot(genfromtxt("./dogscats_results/acc_"+str(optimizer)+".txt"), label = str(optimizer))

plt.title('accuracy')  
plt.ylabel('accuracy')  
plt.xlabel('epoch')  
pyplot.legend();
pyplot.show();
```

```python
plt.figure(figsize=(15, 10))

for optimizer in Optimizers:
    pyplot.plot(genfromtxt("./dogscats_results/val_loss_"+str(optimizer)+".txt"), label = str(optimizer))
        
plt.title('val. loss')  
plt.ylabel('val. loss')  
plt.xlabel('epoch')  
pyplot.legend();
pyplot.show();
```

```python
plt.figure(figsize=(15, 10))

for optimizer in Optimizers:
    pyplot.plot(genfromtxt("./dogscats_results/val_acc_"+str(optimizer)+".txt"), label = str(optimizer))
    
plt.title('val. accuracy')  
plt.ylabel('val. accuracy')  
plt.xlabel('epoch')  
pyplot.legend();
pyplot.show();
```

```python
plt.figure(figsize=(15, 10))

print "Maximum value of Validation Accuracy"
for optimizer in Optimizers:
    print str(optimizer), np.max(np.loadtxt("./dogscats_results/val_acc_"+str(optimizer)+".txt"))
                 
```
